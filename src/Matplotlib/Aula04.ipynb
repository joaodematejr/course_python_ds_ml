```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y="tip", data=tips, hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y="tip", data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s': 100}, col='day', aspect=0.6, size=8)
```
