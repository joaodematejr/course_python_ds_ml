```python
import numpy as np
x = np.arange(0, 100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Titulo')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

```python
ax1.plot(x, y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

fig
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
```

```python
ax.plot(x, z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x, y, color='blue', ls='--', lw=3)
axes[1].plot(x, z, color='red', lw=3)
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))

axes[0].plot(x, y, color='blue', ls='--', lw=3)
axes[1].plot(x, z, color='red', lw=3)
fig
```
